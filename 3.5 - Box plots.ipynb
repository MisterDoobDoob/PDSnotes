### Box Plots
A box plot or boxplot is a method for graphically depicting groups of numerical data through their quartiles (25 Percentiles). <br>
Good for displaying <strong>continuous</strong> data. <br>
A boxplot is a standardized way of displaying the dataset based on a five-number summary: the minimum, the maximum, the sample median, and the first and third quartiles. <br>
(Generally, we can avoid the outliers)

(Actual) Minimum (Q0 or 0th percentile): the lowest data point excluding any outliers.<br>
(Actual) Maximum (Q4 or 100th percentile): the largest data point excluding any outliers.<br>
Median (Q2 or 50th percentile): the middle value of the dataset.<br>
First quartile (Q1 or 25th percentile): also known as the lower quartile qn(0.25), is the median of the lower half of the dataset.<br>
Third quartile (Q3 or 75th percentile): also known as the upper quartile qn(0.75), is the median of the upper half of the dataset.<br>

An important element used to construct the box plot by determining the minimum and maximum data values feasible, but is not part of the aforementioned five-number summary, is the interquartile range or IQR denoted below:
Interquartile range (IQR) : is the distance between the upper and lower quartiles.

    IQR = Q3  Q1 = q n ( 0.75 )  q n ( 0.25 ) {\displaystyle {\text{IQR}}=Q_{3}-Q_{1}=q_{n}(0.75)-q_{n}(0.25)} {\displaystyle {\text{IQR}}=Q_{3}-Q_{1}=q_{n}(0.75)-q_{n}(0.25)}

#### Box Plot Eg 0 - Gaussian Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
gaussian_numbers = np.random.randn(1000)
plt.hist(gaussian_numbers)
plt.title("Gaussian Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

plt.boxplot(gaussian_numbers, vert = 0)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

#### Box Plot Eg 1 - Sleep Distribution


```python
import matplotlib.pyplot as plt
import numpy as np

# No of hours 6 students sleep in a week
data = np.array([(8,5,7,3,4,5,9),(6,5,6,8,8,8,6),(4,5,7,6,6,7,9),
                 (3,5,7,8,8,7,6),(3,6,7,7,7,6,6),(8,7,6,6,7,7,9)])
labels = np.array(['Sun','Mon', 'Tue','Wed','Thu','Fri','Sat'])

plt.boxplot(data,labels=labels)
plt.title("Number of hours 6 students sleep in a week")
plt.show()
```

From the output, we can tell that most of the students do get more than 6 hours of sleep on most days.

Saturday is probably a "rest day", where students get the most sleep. Additionally, their sleep patterns are alot closer together (notice there are no whiskers!)

However, there is a huge disparity of sleep patterns on Sunday where some students get much less sleep than others.

Most students sleep very little on Mondays with little difference.  The sleeping patterns of Tue, and Fri also have little variation. 

Sleep patterns on Thursday vary by quite a significant amount too.

(On a macro scale, having a sample size of 6 is also probably not very representative of the overall student cohort)

#### Box plot E.g. 2 - COE data

```python
# Overlay numeric value of median and outliers in a boxplot
import numpy as np
import matplotlib.pyplot as plt

# read in the dataset
data = np.genfromtxt("data/coe-results.csv",
            delimiter=',',
            names=True, dtype=('U7',int,'U10',int,int,int,int))

# set up the y values
catA = data[data['vehicle_class'] == 'Category A']['quota']
catB = data[data['vehicle_class'] == 'Category B']['quota']
catC = data[data['vehicle_class'] == 'Category C']['quota']
catD = data[data['vehicle_class'] == 'Category D']['quota']
catE = data[data['vehicle_class'] == 'Category E']['quota']
y_values = np.array([catA, catB, catC, catD, catE])

# set up the labels on the x-axis
x_labels = np.unique(data['vehicle_class'])
x_ticks = np.arange(len(x_labels))
plt.xticks(rotation=90)

# plot the box plot
plt.title("Vehicle Quotas per category")
plt.boxplot(y_values.transpose(),labels=x_labels, patch_artist=True)
plt.show()
```

(Some context:)<br>

| Category 	| Detail                                  	|
|:----------:|-----------------------------------------:|
| A        	| Cars with Engine Capacity up to 1,600cc 	|
|   B      	| Cars with engine capacity above 1,600cc 	|
|   C      	| Goods Vehicles and Buses                	|
|   D      	| Motorcycles                             	|
|   E      	| Open (for any other kind of vehicle)    	|

#### Observations:
C and D have a lot of variation, especially lots of outliers.

Category A has the highest median, and generally has the highest quota available
